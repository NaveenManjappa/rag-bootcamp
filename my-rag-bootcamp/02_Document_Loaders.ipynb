```python
import os
from pathlib import Path
from dotenv import load_dotenv

load_dotenv()
print("Environment loaded")
print(f"Current directory:{os.getcwd()}")
print(f"Sample data directory exists: {Path('sample_data').exists()}")
```

```python
from langchain_core.documents import Document
from langchain_community.document_loaders import PyPDFLoader

pdf_path="./Content/resume.pdf"

if Path(pdf_path).exists():
  loader = PyPDFLoader(pdf_path)

  documents = loader.load()
  print(f"Loaded {len(documents)} pages\n")
  print(f"Content: {documents[0].page_content[:200]}")
  print(f"Metadata: {documents[0].metadata}")
```

```python
#Lazy load example
if Path(pdf_path).exists():
  loader = PyPDFLoader(pdf_path)

  for i,page in enumerate(loader.lazy_load(),1):
    if i>3:
      break
    print(f"Page {i}")
    print(f"Length: {len(page.page_content)}")
    print(f"Preview: {page.page_content[:100]}")
```

```python
from langchain_community.document_loaders import CSVLoader
csv_path="sample_data/products.csv"

if Path(csv_path).exists():
  loader = CSVLoader(
    file_path=csv_path,
    source_column="product_name"
  )

  documents = loader.load()

  print(f"Loaded {len(documents)} products")
  for i,doc in enumerate(documents,1):
    print(f"Product: {i}")
    print(doc.page_content)
    print(f"Source: {doc.metadata['source']}")
    print(f"{doc.metadata.get('row','N/A')}")
```

```python
#JSON loader
from langchain_community.document_loaders import JSONLoader

json_path="sample_data/api_response.json"

if Path(json_path).exists():
  loader = JSONLoader(
    file_path=json_path,
    jq_schema=".articles[]",
    text_content=False
  )

  documents = loader.load()
  print(f"Loaded {len(documents)} articles\n")
  print(f"Content:{documents[0].page_content}")
  print(f"Metadat:{documents[0].metadata}")
```
